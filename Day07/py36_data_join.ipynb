#### 판다스 데이터 합치기

- 옆으로 합치기 (merge)
- 위아래로 합치기 (append)

```python
import pandas as pd
import numpy as np
```

```python
# sample_1 데이터 불러오기
sample_1 = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_1.xlsx',
                         sheet_name='Sheet1', header=1, skipfooter=2)
# header를 0에서 1로 변경 / 2019년 11월 입국객 수는 컬럼 헤더가 아님
# skipfooer는 받는 데이터 중에서 합계 등이 있으므로 총합계, 전년동기를 제거
```

```python
sample_1
```

```python
# 코드마스터 데이터 불러오기
code_master = pd.read_excel('https://raw.githubusercontent.com/Play-with-data/datasalon/master/02_%EA%B0%9C%EC%A0%95%ED%8C%90/2_Data_Analysis_Basic/files/sample_codemaster.xlsx')

code_master
```

```python
# merge == DB join
merge_result = pd.merge(left=sample_1
                        ,right=code_master
                        ,how='left'
                        ,left_on='국적코드'
                        ,right_on='국적코드')
merge_result
```

```python
# merge == DB left iner join과 일치
merge_result = pd.merge(left=sample_1
                        ,right=code_master
                        ,how='inner'
                        ,left_on='국적코드'
                        ,right_on='국적코드')
merge_result
```
